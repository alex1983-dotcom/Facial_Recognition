```python
import cv2
import matplotlib.pyplot as plt
from src.face_utils.detector import FaceDetector

# Загрузка изображения
img = plt.imread("../data/raw/test_2.jpg")  # RGB

# Обнаружение
detector = FaceDetector(confidence_threshold=0.6)
boxes = detector.detect(img)

# Визуализация
fig, ax = plt.subplots(1, figsize=(10, 8))
ax.imshow(img)
for (x, y, w, h) in boxes:
    rect = plt.Rectangle((x, y), w, h, linewidth=2, edgecolor='red', facecolor='none')
    ax.add_patch(rect)
ax.axis("off")
plt.show()
```

```python
# InsightFace
from src.face_utils.insightface_detector import InsightFaceDetector

detector = InsightFaceDetector(confidence_threshold=0.4)
boxes = detector.detect(img)
print(f"Найдено лиц: {len(boxes)}")

# Визуализация через matplotlib (без draw_boxes)
fig, ax = plt.subplots(1, figsize=(12, 8))
ax.imshow(img)
for (x, y, w, h) in boxes:
    rect = plt.Rectangle((x, y), w, h, linewidth=2, edgecolor='green', facecolor='none')
    ax.add_patch(rect)
ax.axis("off")
plt.title("InsightFace Detection")
plt.show()
```
